### **Outliers Values**

Pada tahap ini, kita akan mengidentifikasi outlier pada kolom numerik. Hal ini penting agar analisis dan model yang dibangun menghasilkan insight yang akurat.

Namun, tidak semua outlier perlu ditangani. Kita perlu memahami konteks tiap kolom, karena nilai ekstrem bisa jadi justru memberikan insight berharga, misalnya siswa dengan jam belajar sangat tinggi atau latihan soal di atas rata-rata.

Dan terlihat kolom

* Sample Question Papers Practiced

Dengan presentase 0.03% (dibawah 1%)

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

## **Composition/Komposisi**

1. Komposisi Siswa Berdasarkan Kategori Performa

Aktivitas: Mengelompokkan siswa ke dalam 3 kategori performa (Rendah, Sedang, Tinggi) berdasarkan Performance Index dan menampilkan proporsinya.

Tujuan: Mengetahui komposisi performa akademik siswa, apakah mayoritas berada di kategori sedang, atau justru banyak yang performanya rendah/tinggi.

## **Construct Data**

Pada Tahan ini akan dilakukan penambahan kolom status, karena kolom notes tidak mempunyai nilai yang tidak relevan dengan nilai manapun.

```python
def Penilaian (Performance_Index):
    if Performance_Index > 85:
        return 'Luar Biasa'
    elif Performance_Index < 75:
        return 'Kurang'
    else:
        return 'Cukup'

df['Status'] = df['Performance Index'].apply(Penilaian)
```

Acuan dari kolom status ini adalah performance index yang memungkinkan mendapat penilaian yang lebih pasti. Dibagi menjadi 3, luar bisa, kurang dan cukup, dengan begitu akan memudahkan pihak akademik untuk memeriksa perkembangan murid.
