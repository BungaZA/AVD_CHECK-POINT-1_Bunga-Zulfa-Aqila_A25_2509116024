### **Outliers Values**

Pada tahap ini, kita akan mengidentifikasi outlier pada kolom numerik. Hal ini penting agar analisis dan model yang dibangun menghasilkan insight yang akurat.

Namun, tidak semua outlier perlu ditangani. Kita perlu memahami konteks tiap kolom, karena nilai ekstrem bisa jadi justru memberikan insight berharga, misalnya siswa dengan jam belajar sangat tinggi atau latihan soal di atas rata-rata.

Dan terlihat kolom

* Sample Question Papers Practiced

Dengan presentase 0.03% (dibawah 1%)

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

1. Performance Index berdasarkan Ekstrakurikuler

Aktivitas: Membandingkan rata-rata Performance Index antara siswa yang mengikuti ekstrakurikuler dengan yang tidak.

Tujuan: Mengetahui apakah keikutsertaan ekstrakurikuler berdampak positif atau negatif terhadap prestasi akademik.

```python
performa_ekstra = df.groupby('Extracurricular Activities')['Performance Index'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(8,6))
sns.barplot(x=performa_ekstra['Extracurricular Activities'],
            y=performa_ekstra['Performance Index'],
            palette='viridis',
            hue=performa_ekstra['Extracurricular Activities'],
            legend=False)
plt.title('Rata-rata Performance Index Berdasarkan Keikutsertaan Ekstrakurikuler', fontsize=12, fontweight='bold')
plt.xlabel('Ekstrakurikuler')
plt.ylabel('Rata-rata Performance Index')
plt.ylim(0, 100)
plt.show()
```

Tujuan: Menentukan rentang jam belajar paling efektif untuk mencapai performa optimal.

```python
df['Hours Category'] = pd.cut(df['Hours Studied'],
                               bins=[0, 3, 5, 7, 10],
                               labels=['Rendah (0-3 jam)', 'Sedang (4-5 jam)', 'Tinggi (6-7 jam)', 'Sangat Tinggi (8-10 jam)'])

performa_jam = df.groupby('Hours Category')['Performance Index'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x=performa_jam['Hours Category'],
            y=performa_jam['Performance Index'],
            palette='plasma',
            hue=performa_jam['Hours Category'],
            legend=False)
plt.title('Rata-rata Performance Index Berdasarkan Kategori Jam Belajar', fontsize=12, fontweight='bold')
plt.xlabel('Kategori Jam Belajar per Hari')
plt.ylabel('Rata-rata Performance Index')
plt.ylim(0, 100)
plt.show()
```

Insight: Terdapat tren peningkatan performa yang konsisten seiring bertambahnya jam belajar. Siswa dengan jam belajar 8-10 jam mencapai rata-rata 64, jauh lebih tinggi dibanding siswa yang belajar <3 jam (48). ## **Composition/Komposisi**

1. Komposisi Siswa Berdasarkan Kategori Performa

Aktivitas: Mengelompokkan siswa ke dalam 3 kategori performa (Rendah, Sedang, Tinggi) berdasarkan Performance Index dan menampilkan proporsinya.

Tujuan: Mengetahui sebaran tingkat performa akademik siswa, apakah mayoritas berada di kategori sedang, atau justru banyak yang performanya rendah/tinggi.

Visualisasi: Pie Chart

```python
bins = [0, 40, 70, 100]
labels = ['Rendah (0-40)', 'Sedang (41-70)', 'Tinggi (71-100)']
df['Performa_Kategori'] = pd.cut(df['Performance Index'], bins=bins, labels=labels)

performa_count = df['Performa_Kategori'].value_counts()

plt.figure(figsize=(8, 8))
performa_count.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff6b6b', '#ffd93d', '#6bcf7f'])
plt.title('Proporsi Siswa Berdasarkan Kategori Performa Akademik')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Insight: Hampir setengah dari total siswa (49.3%) berada di kategori performa sedang (41-70), sementara siswa berperforma tinggi (71-100) dan rendah (0-40) masing-masing sekitar 25%. Komposisi ini menunjukkan distribusi yang cukup sehat dengan dominasi siswa di level menengah, namun masih ada 25% siswa yang perlu perhatian khusus karena berada di kategori rendah.

2. Komposisi Performa Berdasarkan Status Ekstrakurikuler

Aktivitas: Menampilkan proporsi siswa dengan performa rendah, sedang, dan tinggi, yang dipisah berdasarkan status ekstrakurikuler.

Tujuan: Melihat apakah distribusi performa berbeda antara siswa yang aktif vs tidak aktif ekstrakurikuler.

Visualisasi: Stacked Bar Chart

```python
performa_yes = df[df['Extracurricular Activities'] == 'Yes']['Performa_Kategori'].value_counts()

performa_no = df[df['Extracurricular Activities'] == 'No']['Performa_Kategori'].value_counts()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.pie(performa_yes.values, labels=performa_yes.index, autopct='%1.1f%%',
        startangle=140, colors=['#ff6b6b', '#ffd93d', '#6bcf7f'])
ax1.set_title('Komposisi Performa Siswa Aktif Ekstrakurikuler')
ax1.axis('equal')

ax2.pie(performa_no.values, labels=performa_no.index, autopct='%1.1f%%',
        startangle=140, colors=['#ff6b6b', '#ffd93d', '#6bcf7f'])
ax2.set_title('Komposisi Performa Siswa Tidak Aktif Ekstrakurikuler')
ax2.axis('equal')

plt.suptitle('Perbandingan Komposisi Performa Berdasarkan Status Ekstrakurikuler', fontsize=14)
plt.tight_layout()
plt.show()
```

Insight: Dari siswa yang aktif ekstrakurikuler, proporsi performa tinggi (26.3%) justru sedikit lebih besar dibanding performa rendah (23.8%). Ini memperkuat temuan sebelumnya bahwa ekstrakurikuler tidak berdampak negatif. Mayoritas siswa aktif ekstrakurikuler (49.9%) berada di performa sedang, konsisten dengan komposisi umum.
