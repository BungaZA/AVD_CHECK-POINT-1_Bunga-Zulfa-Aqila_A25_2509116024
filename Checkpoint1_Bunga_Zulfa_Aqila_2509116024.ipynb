# **Business Understanding**

## **Business Objective**

Dataset ini berisi informasi mengenai kebiasaan belajar dan faktor-faktor yang memengaruhi performa akademik siswa. Terdapat 10.005 baris data dan 7 kolom yang berisi jam belajar, nilai sebelumnya, aktivitas ekstrakurikuler, jam tidur, jumlah soal latihan yang dikerjakan, serta indeks performa siswa.

Tujuan analisis ini adalah untuk mencari tahu kebiasaan belajar apa yang paling berpengaruh terhadap nilai siswa. Hasilnya nanti bisa dipakai sekolah atau guru untuk membantu siswa yang mengalami kesulitan belajar.

## **Assess Situation**

Setiap siswa punya kebiasaan belajar yang berbeda-beda. Ada yang rajin belajar tapi kurang tidur, sibuk ekstrakurikuler sampai jarang latihan soal, dan belajarnya sebentar tapi nilainya bagus.

Tantangannya adalah bagaimana menemukan pola kebiasaan yang sehat dan efektif untuk meningkatkan performa akademik. Dari dataset ini, kita bisa melihat hubungan antara kebiasaan sehari-hari siswa dengan nilai yang mereka dapatkan. Data ini bisa jadi petunjuk buat guru dalam mendampingi siswa agar bisa belajar lebih optimal.

## **Analytic Goals**

Tujuan Analisis:

1. Menganalisis hubungan antara jam belajar dan performa siswa.
2. Mengukur pengaruh jam tidur terhadap performa akademik.
3. Mengetahui apakah latihan soal berpengaruh signifikan terhadap nilai.
4. Membandingkan performa siswa yang mengikuti ekstrakurikuler dan yang tidak.
